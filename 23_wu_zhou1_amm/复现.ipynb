- **作者**：王凯

#### 标题


本文用一种**事件触发**的控制方法，去解决**半马尔可夫切换系统**的稳定性问题，并且要满足**扩展耗散性**性能指标


#### 摘要


- **不确定性**和**外部扰动**是在现实中必须考虑的因素.
- 提出了新的、改进的**事件触发**规则，它不仅依赖于系统当前的状态，还同时考虑了**系统所处的模式**和**外部扰动**的大小，因此能比现有的方法更高效地减少不必要的触发。
- 为了在数学上证明方法有效，构建了一个特殊的**李雅普诺夫泛函**——一个非常巧妙的、分段定义的**能量函数**，通过证明这个函数最终是收敛的，来严格证明整个系统是稳定的。
- 通过求解线性矩阵不等式（LMIs），同时计算出**控制器参数**和**事件触发器参数**。
- 通过两个实际的例子（直流电机和机械臂）进行了仿真，验证了提出的理论和方法的有效性。


#### 预备知识

##### 系统的数学模型

**不确定半马尔可夫切换系统**的通用数学模型：

$$
\begin{aligned}
	\dot{x}\left( t \right) &=\bar{A}\left( \zeta \left( t \right) \right) x\left( t \right) +B\left( \zeta \left( t \right) \right) u\left( t \right) +C\left( \zeta \left( t \right) \right) \omega \left( t \right)\\
	y\left( t \right) &=D\left( \zeta \left( t \right) \right) x\left( t \right)\\
	x\left( 0 \right) &=0,\zeta \left( 0 \right) =\zeta _0\\
\end{aligned}
$$


**详细解释**：

- $x(t)$：系统状态向量
- $u(t)$：控制输入（由我们设计，以稳定系统）
- $y(t)$：系统测量输出
- $\omega(t)$：作用于系统的外部扰动
- $\zeta(t)$：半马尔科夫过程，决定当前系统所处模式
- $\bar{A},B,C,D$：系数矩阵，依赖于当前模式$\zeta(t)$。其中，$\bar{A}=A+\Delta A$


##### 三个基本假设

- **假设 1**: 系统是**可控的** 。

  - **解释**: 这意味着控制输入 $u(t)$ 确实能够有效地影响系统的所有状态。如果一个系统不可控，那任何控制器都将无能为力。

- **假设 2**: 外部扰动是**可测量的** 。

  - **解释**: 这是本文一个比较强的假设。它意味着控制器不仅知道系统的状态，还能实时知道外部扰动的大小。这一点对于设计扰动依赖的触发器至关重要。

- **假设 3**: 参数不确定性是**范数有界的** 。
  - **解释**: 这意味着虽然不知道不确定项 $\Delta A$ 的精确值，但知道它的大小在一个已知的范围内，不会无限大 。这为设计一个能应对这种不确定性的鲁棒控制器提供了可能性。


##### 控制器的数学形式

$$u(t) = K(\zeta(t))x(t_s), \quad t \in [t_s, t_{s+1})$$

**详细解释**：

- **$K(\zeta(t))$**: **控制器增益矩阵**。
- **$x(t_s)$**: 控制器在计算当前控制量 $u(t)$ 时，使用的**不是**系统当前时刻的实时状态 $x(t)$，而是上一次触发时刻 $t_s$ 的状态。
- **$t \in [t_s, t_{s+1})$**: 这个时间区间表示，从上一次触发时刻 $t_s$ 开始，直到下一次触发时刻 $t_{s+1}$ 到来之前，控制器的输出 $u(t)$ 都是基于旧的状态信息 $x(t_s)$ 计算的，并且其值保持不变。


##### 事件触发器的数学形式

$$
t_{s+1}=\min\text{\{}t\ge t_s+d\mid \left( x\left( t \right) -x\left( t_s \right) \right) ^T\Lambda \left( \zeta \left( t \right) \right) \left( x\left( t \right) -x\left( t_s \right) \right) \ge \sigma x^T\left( t \right) \Lambda \left( \zeta \left( t \right) \right) x\left( t \right) +\delta \omega ^T\left( t \right) \omega \left( t \right) \}
$$


**详细解释**：
加权的误差能量 $\ge$ 动态的触发阈值

- $\sigma x^T(t)\Lambda_i x(t) + \delta \omega^T(t)\omega(t)$：取决于当前状态 $x(t)$ 的幅度和外部扰动 $\omega(t)$ 的幅度


##### 目标一：随机稳定性

**定义**：一个系统被称为随机稳定，指的是：在没有外部扰动 $\omega(t)=0$ 的情况下，系统状态 $x(t)$ 随时间演变的轨迹的总“能量”（用状态向量的范数平方 $||x(t)||^2$ 来衡量），其数学期望（即在所有可能的随机切换路径下的平均值）是一个有限的数值。


即：$$\mathcal{E}\left\{\int_{0}^{\infty} ||x(\mu)||^2 d\mu\right\} \le M_{\circ}$$
这意味着，即使系统内部的模式在随机跳跃，只要没有外部干扰，控制器也必须能保证系统状态最终会回归到 0。从能量角度看，系统从初始状态出发所产生的总能量波动，平均来说，是有限的，而不是发散到无穷大。


##### 目标二：扩展耗散性

$$\sup_{t \ge 0}\{\mathcal{E}(x^T(t)\Upsilon_4 x(t))\} \le \mathcal{E}\left\{\int_0^t J(\mu)d\mu\right\}$$
其中，$$J(t)=x^T(t)\Upsilon_1x(t)+2x^T(t)\Upsilon_2\omega(t)+\omega^T(t)\Upsilon_3\omega(t)$$
